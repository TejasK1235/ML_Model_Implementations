```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import StackingRegressor,VotingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
import sklearn
```

```python
data = sklearn.datasets.load_diabetes()
```

```python
X = pd.DataFrame(data.data,columns=data.feature_names)
y = pd.Series(data.target,name="target")
y = pd.DataFrame(y)
```

```python
X.shape
```

```python
y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=24)
```

```python
knn = KNeighborsRegressor()
dtr = DecisionTreeRegressor()
rf = RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
voting = VotingRegressor(estimators=[('knn',knn),('dtr',dtr),('rf',rf)])
```

```python
voting.fit(X_train,y_train)
```

```python
voting_preds = voting.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(voting_preds,y_test)
```

```python
r2_score(voting_preds,y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
stacking = StackingRegressor(estimators=[('knn',knn),('dtr',dtr),('rf',rf)],final_estimator=LinearRegression())
```

```python
stacking.fit(X_train,y_train)
```

```python
stacking_preds = stacking.predict(X_test)
```

```python
mean_squared_error(stacking_preds,y_test)
```

```python
r2_score(stacking_preds,y_test)
```

```python
gb = GradientBoostingRegressor(n_estimators=100,learning_rate=0.01,random_state=42)
xgb = XGBRegressor(n_estimators=100,learning_rate=0.01,random_state=42)
ada = AdaBoostRegressor(n_estimators=100,learning_rate=0.01,random_state=42)
```

```python
gb.fit(X_train,y_train)
```

```python
xgb.fit(X_train,y_train)
```

```python
ada.fit(X_train,y_train)
```

```python
gb_preds = gb.predict(X_test)
xgb_preds = xgb.predict(X_test)
ada_preds = ada.predict(X_test)
```

```python
print(f"MSE of GB: {mean_squared_error(gb_preds,y_test)}")
print(f"MSE of XGB: {mean_squared_error(xgb_preds,y_test)}")
print(f"MSE of ADA: {mean_squared_error(ada_preds,y_test)}")
```

```python
print(f"R2_Score of GB: {r2_score(gb_preds,y_test)}")
print(f"R2_Score of XGB: {r2_score(xgb_preds,y_test)}")
print(f"R2_Score of ADA: {r2_score(ada_preds,y_test)}")
```
